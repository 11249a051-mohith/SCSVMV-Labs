# Topic 3: Functions in R

```r
# Q1: Bank Interest Calculator
simple_interest <- function(p, r, t){
  si <- (p * r * t) / 100
  return(si)
}
simple_interest(10000, 5, 2)
```

```r
# Q2: Student Result Processing
check_result <- function(marks){
  if(marks >= 40){
    return("Pass")
  } else {
    return("Fail")
  }
}
check_result(65)
```

```r
# Q3: Employee Salary Calculation
net_salary <- function(basic_salary){
  tax <- basic_salary * 0.10
  net <- basic_salary - tax
  return(net)
}
net_salary(30000)
```

```r
# Q4: Electricity Bill Generator
electricity_bill <- function(units){
  if(units <= 100){
    bill <- units * 2
  } else if(units <= 200){
    bill <- (100 * 2) + (units - 100) * 3
  } else {
    bill <- (100 * 2) + (100 * 3) + (units - 200) * 5
  }
  return(bill)
}
electricity_bill(250)
```

```r
# Q5: Online Shopping Discount
final_amount <- function(amount){
  if(amount >= 5000){
    discount <- amount * 0.20
  } else if(amount >= 2000){
    discount <- amount * 0.10
  } else {
    discount <- 0
  }
  return(amount - discount)
}
final_amount(4500)
```

```r
# Q6: Temperature Conversion System
celsius_to_fahrenheit <- function(c){
  f <- (c * 9/5) + 32
  return(f)
}
celsius_to_fahrenheit(37)
```

```r
# Q7: Grading System
assign_grade <- function(marks){
  if(marks >= 90){
    "A"
  } else if(marks >= 75){
    "B"
  } else if(marks >= 50){
    "C"
  } else {
    "Fail"
  }
}
assign_grade(88)
```

```r
# Q8: Voting Eligibility Checker
check_voting <- function(age){
  if(age >= 18){
    return("Eligible to Vote")
  } else {
    return("Not Eligible to Vote")
  }
}
check_voting(17)
```

```r
# Q9: Bank Loan EMI Calculator
calculate_emi <- function(principal, rate, years){
  monthly_rate <- rate / (12 * 100)
  months       <- years * 12
  emi <- (principal * monthly_rate * (1 + monthly_rate)^months) /
         ((1 + monthly_rate)^months - 1)
  return(round(emi, 2))
}
calculate_emi(500000, 7.5, 10)
```

```r
# Q10: Hospital Patient Billing System
hospital_bill <- function(consultation, room_charge, days){
  subtotal <- consultation + (room_charge * days)
  tax      <- subtotal * 0.05
  total    <- subtotal + tax
  return(total)
}
hospital_bill(500, 2000, 3)
```

```r
# Q11: Online Shopping Order Processing System
order_total <- function(cart_value){
  if(cart_value >= 3000){
    discount <- cart_value * 0.10
  } else {
    discount <- 0
  }
  if(cart_value >= 5000){
    delivery <- 0
  } else {
    delivery <- 100
  }
  final_amount <- cart_value - discount + delivery
  return(final_amount)
}
order_total(4200)
```

```r
# Q12: University GPA Calculation System
calculate_gpa <- function(marks){
  avg <- mean(marks)
  if(avg >= 85){
    result <- "Distinction"
  } else if(avg >= 70){
    result <- "First Class"
  } else if(avg >= 50){
    result <- "Second Class"
  } else {
    result <- "Fail"
  }
  return(result)
}
calculate_gpa(c(78, 85, 90, 72, 80))
```

```r
# Q13: Smart Electricity Meter System
smart_meter_bill <- function(peak_units, offpeak_units){
  bill        <- (peak_units * 6) + (offpeak_units * 3)
  total_units <- peak_units + offpeak_units
  if(total_units > 300){
    bill <- bill * 1.10
  }
  return(bill)
}
smart_meter_bill(180, 150)
```

```r
# Q14: Insurance Premium Calculation
insurance_premium <- function(base, age, smoker){
  premium <- base
  if(age > 50){
    premium <- premium * 1.20
  }
  if(smoker){
    premium <- premium * 1.30
  }
  return(premium)
}
insurance_premium(5000, 55, TRUE)
```

```r
# Q15: Employee Payroll Processing
payroll <- function(basic){
  hra   <- basic * 0.20
  da    <- basic * 0.10
  gross <- basic + hra + da
  tax   <- gross * 0.12
  net   <- gross - tax
  return(net)
}
payroll(40000)
```

```r
# Q16: Smart City Water Usage Monitoring
water_usage_status <- function(litres){
  if(litres <= 500){
    "Normal"
  } else if(litres <= 1000){
    "High Usage"
  } else {
    "Critical"
  }
}
water_usage_status(1200)
```

# Topic 4: Data Frames in R (cbind, rbind, merge)

```r
# Q1: College Student Record System
students <- data.frame(
  RollNo = c(101, 102, 103),
  Name   = c("Ravi", "Meena", "Karthik"),
  Dept   = c("CSE", "ECE", "IT")
)

marks <- data.frame(Marks = c(78, 85, 90))

student_data <- cbind(students, marks)
print(student_data)

new_students <- data.frame(
  RollNo = 104, Name = "Anjali", Dept = "CSE", Marks = 88
)

final_data <- rbind(student_data, new_students)
print(final_data)
```

# Topic 7: Read CSV File and Analyze Data in R

This topic creates a CSV file first, then reads and analyzes it.

```r
# Step 0: Create the sales_data.csv file
sales_raw <- data.frame(
  Date     = c("2026-01-01","2026-01-01","2026-01-02","2026-01-02","2026-01-03","2026-01-03"),
  Product  = c("Phone","Bag","Laptop","Headset","Bag","Phone"),
  Category = c("Electronics","Accessories","Electronics","Electronics","Accessories","Electronics"),
  Quantity = c(5, 10, 3, 7, 5, 3),
  Price    = c(20000, 1500, 50000, 2500, 1500, 20000)
)
write.csv(sales_raw, "sales_data.csv", row.names=FALSE)
cat("CSV file created!\n")
```

```r
# Step 1: Load the CSV File
sales <- read.csv("sales_data.csv")
head(sales)
str(sales)
summary(sales)
```

```r
# Step 2: Data Cleaning
sales$Product  <- trimws(sales$Product)
sales$Category <- trimws(sales$Category)
sales$Date     <- as.Date(sales$Date, format="%Y-%m-%d")
sales$TotalSales <- sales$Quantity * sales$Price
head(sales)
```

```r
# Step 3: Total Sales per Product
total_sales_product <- aggregate(TotalSales ~ Product, data=sales, sum)
print(total_sales_product)
```

```r
# Step 4: Product with Highest Sales
top_product <- total_sales_product[which.max(total_sales_product$TotalSales), ]
print(top_product)
```

```r
# Step 5: Total Quantity Sold per Category
quantity_per_category <- aggregate(Quantity ~ Category, data=sales, sum)
print(quantity_per_category)
```

```r
# Step 6: Filter High-Value Products (>10000)
high_sales <- total_sales_product[total_sales_product$TotalSales > 10000, ]
print(high_sales)
```

# Topic 8: Pie Charts and Bar Charts in R

```r
# Q1: University Course Enrollment
departments <- c("CSE", "ECE", "IT", "Mechanical")
students    <- c(120, 80, 60, 40)

pie(
  students,
  labels = departments,
  main   = "Student Enrollment per Department",
  col    = c("skyblue", "orange", "green", "pink")
)

barplot(
  students,
  names.arg = departments,
  main      = "Number of Students per Department",
  xlab      = "Department",
  ylab      = "Number of Students",
  col       = c("skyblue", "orange", "green", "pink")
)
```

```r
# Q2: Retail Store Sales Distribution
category <- c("Electronics", "Clothing", "Accessories", "Home Appliances")
sales    <- c(150000, 75000, 50000, 25000)

pie(
  sales,
  labels = category,
  main   = "Sales Distribution by Category",
  col    = rainbow(length(category))
)

barplot(
  sales,
  names.arg = category,
  main      = "Sales per Category",
  xlab      = "Category",
  ylab      = "Sales (Rs.)",
  col       = rainbow(length(category))
)
```

```r
# Q3: Online Exam Result Analysis
grades <- c("A", "B", "C", "D")
count  <- c(15, 25, 30, 10)

pie(
  count,
  labels = grades,
  main   = "Grade Distribution",
  col    = c("gold", "lightblue", "pink", "lightgreen")
)

barplot(
  count,
  names.arg = grades,
  main      = "Number of Students per Grade",
  xlab      = "Grade",
  ylab      = "Number of Students",
  col       = c("gold", "lightblue", "pink", "lightgreen")
)
```

# Topic 9: Statistical Analysis on Dataset in R

```r
# Q1: School Student Marks Analysis
students <- data.frame(
  Name    = c("Arjun","Meena","Rahul","Anitha","Kiran","Sneha","Ravi","Gayathri","Suresh","Lakshmi"),
  Maths   = c(85,90,70,95,80,88,75,92,68,86),
  Science = c(78,88,60,92,85,80,70,90,75,82),
  English = c(92,80,75,98,88,85,78,91,80,87)
)

mean_marks   <- sapply(students[ ,2:4], mean)
median_marks <- sapply(students[ ,2:4], median)
sd_marks     <- sapply(students[ ,2:4], sd)

students$Total <- rowSums(students[ ,2:4])
top_student    <- students[which.max(students$Total), ]

avg_math       <- mean(students$Maths)
above_avg_math <- students[students$Maths > avg_math, ]

mean_marks
median_marks
sd_marks
print(top_student)
print(above_avg_math)
```

```r
# Q2: Retail Sales Data Analysis
sales <- data.frame(
  Product = c("Phone","Laptop","Headset","Bag","Camera","Mouse","Keyboard"),
  Day1    = c(5,3,7,10,2,8,6),
  Day2    = c(4,2,8,12,3,7,5),
  Day3    = c(6,4,6,8,1,9,7),
  Day4    = c(5,3,7,11,2,6,6),
  Day5    = c(7,5,5,9,3,8,8)
)

sales_stats <- data.frame(
  Product  = sales$Product,
  Mean     = apply(sales[ ,2:6], 1, mean),
  Max      = apply(sales[ ,2:6], 1, max),
  Min      = apply(sales[ ,2:6], 1, min),
  Variance = apply(sales[ ,2:6], 1, var)
)

max_avg_product <- sales_stats[which.max(sales_stats$Mean), ]
print(sales_stats)
print(max_avg_product)
```

```r
# Q3: Hospital Patient Age Analysis
patients <- data.frame(
  PatientID = 201:210,
  Name      = c("Suresh","Lakshmi","Arjun","Meena","Ravi","Anitha","Kiran","Sneha","Gayathri","Ravi2"),
  Age       = c(45,38,52,29,60,48,55,35,42,50)
)

mean_age       <- mean(patients$Age)
median_age     <- median(patients$Age)
range_age      <- range(patients$Age)
sd_age         <- sd(patients$Age)
older_patients <- patients[patients$Age > 50, ]

mean_age
median_age
range_age
sd_age
print(older_patients)
```

```r
# Q4: Employee Salary Analysis
employees <- data.frame(
  EmpID  = 101:110,
  Name   = c("Rahul","Sneha","Kiran","Anitha","Ravi","Gayathri","Suresh","Lakshmi","Arjun","Meena"),
  Dept   = c("IT","HR","Finance","IT","HR","Finance","IT","HR","Finance","IT"),
  Salary = c(30000,35000,40000,32000,36000,42000,31000,37000,45000,33000)
)

dept_stats <- aggregate(Salary ~ Dept, data=employees, function(x) c(Mean=mean(x), Median=median(x)))
dept_stats <- do.call(data.frame, dept_stats)

overall_avg  <- mean(employees$Salary)
high_earners <- employees[employees$Salary > overall_avg, ]

print(dept_stats)
print(high_earners)
```

```r
# Q5: Online Store Customer Ratings
ratings <- data.frame(
  ProductID = 1:10,
  Product   = c("Phone","Laptop","Headset","Bag","Camera","Mouse","Keyboard","Charger","Tablet","Speaker"),
  Rating    = c(4,5,3,4,5,4,3,5,4,4)
)

mean_rating   <- mean(ratings$Rating)
median_rating <- median(ratings$Rating)
var_rating    <- var(ratings$Rating)

getmode <- function(v){
  uniqv <- unique(v)
  uniqv[which.max(tabulate(match(v, uniqv)))]
}
mode_rating <- getmode(ratings$Rating)

top_rated <- ratings[ratings$Rating > 4, ]

mean_rating
median_rating
mode_rating
var_rating
print(top_rated)
```
